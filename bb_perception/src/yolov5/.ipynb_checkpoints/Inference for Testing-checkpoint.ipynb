# Testing on Videos

### Load Model

```python
from utils.torch_utils import select_device, load_classifier, time_synchronized
from models.experimental import attempt_load
from utils.general import check_img_size, non_max_suppression, apply_classifier, scale_coords, xyxy2xywh, \
    strip_optimizer, set_logging, increment_path

device = select_device('')
half = device.type != 'cpu'  # half precision only supported on CUDA

# Load model
imgsz = 224
model = attempt_load(r'runs/train/exp4/weights/best.pt', map_location=device)
imgsz = check_img_size(imgsz, s=model.stride.max())
if half:
    model.half()
```

### Define inference method

```python
from utils.datasets import letterbox
from utils.plots import plot_one_box
import random
import torch

conf_thres = 0.5
iou_thres = 0.5

# Get names and colors
names = model.module.names if hasattr(model, 'module') else model.names
colors = [[random.randint(0, 255) for _ in range(3)] for _ in names]

def detect(img0):

    # Run inference
    img = torch.zeros((1, 3, imgsz, imgsz), device=device)  # init img
    _ = model(img.half() if half else img) if device.type != 'cpu' else None  # run once
    
    # Pre-process image
    img = letterbox(img0, new_shape=imgsz)[0]

    # Convert
    img = img[:, :, ::-1].transpose(2, 0, 1)  # BGR to RGB, to 3x416x416
    img = np.ascontiguousarray(img)
    
    img = torch.from_numpy(img).to(device)
    img = img.half() if half else img.float()  # uint8 to fp16/32
    img /= 255.0  # 0 - 255 to 0.0 - 1.0
    if img.ndimension() == 3:
        img = img.unsqueeze(0)

    # Inference
    pred = model(img, augment=False)[0]

    # Apply NMS
    pred = non_max_suppression(pred, conf_thres, iou_thres)

    # Process detections
    for i, det in enumerate(pred):  # detections per image
        gn = torch.tensor(img0.shape)[[1, 0, 1, 0]]  # normalization gain whwh
        if len(det):
            # Rescale boxes from img_size to im0 size
            det[:, :4] = scale_coords(img.shape[2:], det[:, :4], img0.shape).round()

            # Write results
            for *xyxy, conf, cls in reversed(det):
                label = '%s %.2f' % (names[int(cls)], conf)
                plot_one_box(xyxy, img0, label=label, color=colors[int(cls)], line_thickness=3)

    # Return results
    return img0, pred
```

### Run Inference on Single Image

Try on a random image from samples folder

```python
%matplotlib inline
import cv2
import glob
import numpy as np

import matplotlib
import matplotlib.pyplot as plt

is_ipython = 'inline' in matplotlib.get_backend()
if is_ipython:
    from IPython import display
    
plt.ion()

# Get all images in validation images folder
all_images = glob.glob(r'samples/*')

# Get random image from list
rng = np.random.randint(0, len(all_images))
im = cv2.imread(all_images[rng])

disp, pred = detect(im)
print(pred)

plt.imshow(cv2.cvtColor(disp, cv2.COLOR_BGR2RGB));
plt.axis('off');
```

### Run Live Inference

```python
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

while(cap.isOpened()):
    ret, frame = cap.read()

    disp, pred = detect(frame)

    cv2.imshow('frame',disp)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
